```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../data/secondary_clear.xlsx')
```

```python
df.head()
```

```python
chaoxiangNum = df['朝向'].value_counts()
chaoxiang_type = list(chaoxiangNum.index)
chaoxiang_counts = chaoxiangNum.tolist()
print(chaoxiangNum)
print(chaoxiang_type)
print(chaoxiang_counts)
```

```python
xiaoquNum = df['小区名字'].value_counts()
xiaoqu_type = list(xiaoquNum.index)[:18]
xiaoqu_counts = xiaoquNum.tolist()[:18]
print(xiaoquNum)
print(xiaoqu_type)
print(xiaoqu_counts)
```

```python
priceNum = df['每平方（元）'].groupby(df['小区名字']).mean().reset_index().sort_values('每平方（元）', ascending=False).reset_index(drop=True)
price_names = priceNum['小区名字'].tolist()[:10]
price_counts = priceNum['每平方（元）'].tolist()[:10]
print(priceNum)
print(price_names)
print(price_counts)
```

```python
moneyNum = df['总价'].groupby(df['小区名字']).mean().reset_index().sort_values('总价', ascending=False).reset_index(drop=True)
money_names = moneyNum['小区名字'].tolist()[:10]
money_counts = moneyNum['总价'].tolist()[:10]
print(moneyNum)
print(money_names)
print(money_counts)
```

```python
areaNum = df['面积'].groupby(df['小区名字']).mean().reset_index().sort_values('面积', ascending=False).reset_index(drop=True)
area_names = areaNum['小区名字'].tolist()[:10]
area_counts = areaNum['面积'].tolist()[:10]
print(areaNum)
print(area_names)
print(area_counts)
```

```python
timeNum = df['建造时间'].value_counts()
timeSx = df['建造时间']
print(timeNum)
print(timeSx)
```

```python
chengquNum = df['城区'].value_counts()
chengqu_type = list(chengquNum.index)
chengqu_counts = chengquNum.tolist()
print(chengquNum)
print(chengqu_type)
print(chengqu_counts)
```

```python
loucengNum = df['楼层位置'].value_counts()
louceng_type = list(loucengNum.index)
louceng_counts = loucengNum.tolist()

print(loucengNum)
print(louceng_type)
print(louceng_counts)
```
