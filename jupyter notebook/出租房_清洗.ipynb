```python
import numpy as np
import pandas as pd
import csv
```

```python
df = pd.read_csv('../data/rent.csv')
```

```python
df.head()
```

```python
df = df.drop(['mode'], axis=1)
```

```python
df.head()
```

```python
df = df.rename(columns={'community':'楼盘','money':'月租/元','place':'城区','type1':'面积/m²'})
```

```python
df["居室"]=0
df.head(10)
```

```python
centent1 = []
for i in range(len(df)):
    one = df['面积/m²'][i].replace("\xa0","").split(' ')
    o = one[0]
    centent1.append(o)
df['居室'] = centent1
```

```python
df.head()
```

```python
centent2 = []
for i in range(len(df)):
    a = df['面积/m²'][i].replace("\xa0","").split(' ')
    s = a[1]
    centent2.append(s)
df['面积/m²'] = centent2
# print(centent2)
```

```python
df.head()
```

```python
df.to_csv('ren_clear.csv')
```
