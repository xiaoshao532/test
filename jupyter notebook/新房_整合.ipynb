```python
import pandas as pd
```

```python
df = pd.read_csv('../data/new_clear.csv')
```

```python
moneyNum = df['每平方/元'].groupby(df['城区']).mean().reset_index().sort_values('每平方/元', ascending=False).reset_index(drop=True)
money_names = moneyNum['城区'].tolist()
money_counts = moneyNum['每平方/元'].tolist()
print(moneyNum)
print(money_names)
print(money_counts)
```

```python
# 饼图
status = df['销售状态'].value_counts()
status_type = list(status.index)
status_counts = status.tolist()
print(status)
print(status_type)
print(status_counts)
```

```python
# 柱形图
moneyNum = df['每平方/元'].groupby(df['小区']).mean().reset_index().sort_values('每平方/元', ascending=False).reset_index(drop=True)
money_names = moneyNum['小区'].tolist()[0:200:5]
money_counts = moneyNum['每平方/元'].tolist()[0:200:5]
print(moneyNum)
print(money_names)
print(money_counts)
```

```python
# 玫瑰图
chengquNum = df['城区'].value_counts()
chengqu_type = list(chengquNum.index)
chengqu_counts = chengquNum.tolist()
print(chengquNum)
print(chengqu_type)
print(chengqu_counts)
```
