```python
df = df.rename(columns={'community':'楼盘','money':'月租/元','place':'城区','type1':'面积/m²'})
```

```python
df["居室"]=0
df.head(10)
```

```python
centent1 = []
for i in range(len(df)):
    one = df['面积/m²'][i].replace("\xa0","").split(' ')
    o = one[0]
    centent1.append(o)
df['居室'] = centent1
```
