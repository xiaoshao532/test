```python
import pandas as pd
```

```python
df = pd.read_excel('../data/ren_clear2.xlsx')
```

```python
# 饼图
loupanNum = df['楼盘'].value_counts()
loupan_type = list(loupanNum.index)[:16]
loupan_counts = loupanNum.tolist()[:16]
print(loupanNum)
print(loupan_type)
print(loupan_counts)
```

```python
# 柱形图（）
money = df['月租/元'].value_counts()
money_type = list(money.index)[:10]
money_counts = money.tolist()[:10]
print(money)
print(money_type)
print(money_counts)
```

```python
# 地区出租房占比（环图）
place = df['城区'].value_counts()
place_type = list(place.index)
place_counts = place.tolist()
print(place)
print(place_type)
print(place_counts)
```

```python
# 散点图
mianji_sandian = df['面积/m²'].tolist()
money_sandian = df['月租/元'].tolist()
print(mianji_sandian,money_sandian)
```
